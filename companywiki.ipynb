```python
loader1 = TextLoader("/content/leavepolicy.txt")
loader2 = TextLoader("/content/employeebenefits.txt")
loader3 = TextLoader("/content/fakedoc.txt")
loader4 = TextLoader("/content/falseinfo.txt")
docs1 = loader1.load()
docs2 = loader2.load()
docs3 = loader3.load()
docs4= loader4.load()

documents = docs1 + docs2 + docs3 +docs4

print("Documents loaded:", len(documents))
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(documents)

print("Number of chunks:", len(chunks))
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
)

test_embedding = embeddings.embed_query("hello world")

print("Embedding length:", len(test_embedding))
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
)

test_embedding = embeddings.embed_query("hello world")

print("Embedding length:", len(test_embedding))
```

```python
vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="/content/chroma_db"
)
```

```python
retriever = vectorstore.as_retriever()
```

```python
llm = ChatGoogleGenerativeAI(
    model="models/gemini-3.5-flash",
    temperature=0
)
```

```python
query = "what is the name of the company"

retrieved_docs = retriever.invoke(query)

context = "\n".join([doc.page_content for doc in retrieved_docs])

prompt = f"""
Answer the question based only on the context below.

Context:
{context}

Question:
{query}
"""

response = llm.invoke(prompt)

#output = response.content
clean_text = response.content[0]["text"]
print(clean_text)
#print(output)
```
